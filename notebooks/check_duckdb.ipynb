```python
import psycopg2
from psycopg2 import Error
import os
from dotenv import load_dotenv
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
load_dotenv()
try:
    # Kết nối đến PostgreSQL
    connection = psycopg2.connect(
        host=os.getenv("POSTGRES_HOST"),
        database=os.getenv("POSTGRES_DB"),
        user=os.getenv("POSTGRES_USER"),
        password=os.getenv("POSTGRES_PASSWORD"),
        port=os.getenv("POSTGRES_PORT")
    )
    
    # Tạo cursor để thực thi câu lệnh SQL
    cursor = connection.cursor()
    # cursor.execute(f"SELECT DISTINCT education_level FROM jobs")
    cursor.execute("SELECT DISTINCT education_level FROM jobs ORDER BY education_level DESC")
    record = cursor.fetchall()   
    display(pd.DataFrame(record,columns=["education_level"]))    
        
    # Đóng cursor và connection
    cursor.close()
    connection.close()
    print("Đã đóng kết nối PostgreSQL")
    
except (Exception, Error) as error:
    print("Lỗi khi kết nối đến PostgreSQL:", error)
```
